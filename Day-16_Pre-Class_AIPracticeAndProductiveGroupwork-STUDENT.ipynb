```python
# I used claude.ai on March 19, 2023 and my prompt was "can you generate code with comments explaining it" after the first prompt

# ─── Imports ───────────────────────────────────────────────────────────────────
import pandas as pd                          # Data manipulation (not used directly here, but common in data pipelines)
import numpy as np                           # Numerical computing and array operations
from sklearn.preprocessing import StandardScaler    # For normalizing feature values
from sklearn.model_selection import train_test_split # For splitting data into train/test sets
from sklearn.linear_model import LogisticRegression  # Our classification model
from sklearn.metrics import accuracy_score           # For evaluating model performance


# ─── Reproducibility ───────────────────────────────────────────────────────────
np.random.seed(42)  # Setting a seed means random operations produce the same result every run


# ─── Generate Artificial Data ──────────────────────────────────────────────────
n_samples = 1000    # Number of data points (rows)
n_features = 20     # Number of input variables (columns)

# Create a 1000x20 matrix of random floats between 0 and 1
# This represents our "feature" data (like measurements or survey responses)
X = np.random.rand(n_samples, n_features)


# ─── Create Target Labels ──────────────────────────────────────────────────────
# Generate 20 random coefficients — one per feature
# These simulate the "true" relationship between features and the outcome
coefs = np.random.randn(n_features)

# Compute a weighted sum of each row's features using the coefficients
# If the sum is positive → label 1, if negative → label 0
# np.dot performs matrix multiplication: (1000x20) · (20,) → (1000,)
y = np.dot(X, coefs) > 0

# Convert boolean array (True/False) to integers (1/0)
y = y.astype(int)


# ─── Train / Test Split ────────────────────────────────────────────────────────
# Hold out 30% of data for testing; the model never sees this during training
# random_state=42 ensures the same split every run
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42
)


# ─── Feature Scaling ───────────────────────────────────────────────────────────
scaler = StandardScaler()

# fit_transform on training data: learns the mean & std, then scales
X_train = scaler.fit_transform(X_train)

# transform only on test data: applies the SAME scaling learned from training
# (never fit on test data — that would leak information about unseen samples)
X_test = scaler.transform(X_test)


# ─── Train the Model ───────────────────────────────────────────────────────────
model = LogisticRegression()

# The model learns the relationship between features (X_train) and labels (y_train)
model.fit(X_train, y_train)


# ─── Make Predictions ──────────────────────────────────────────────────────────
# Use the trained model to predict labels for the unseen test set
y_pred = model.predict(X_test)


# ─── Evaluate Performance ──────────────────────────────────────────────────────
# Compare predicted labels vs. actual labels
# accuracy_score = number of correct predictions / total predictions
accuracy = accuracy_score(y_test, y_pred)

print('Accuracy:', accuracy)
# A value close to 1.0 means the model predicted most samples correctly
# Since labels were generated from a linear rule, logistic regression should do well here
```

```python
# I used claude.ai on March 19, 2023
# Prompt 1:
#     can you create a version of this code which uses a different library: [code]
# Prompt 2:
#     Q: Which library should replace scikit-learn?
#     A: PyTorch

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# --- Reproducibility ---
np.random.seed(42)
torch.manual_seed(42)

# --- Generate artificial data ---
n_samples = 1000
n_features = 20

X = np.random.rand(n_samples, n_features)
coefs = np.random.randn(n_features)
y = (np.dot(X, coefs) > 0).astype(int)

# --- Train/test split (manual, no sklearn) ---
split = int(0.7 * n_samples)
X_train_np, X_test_np = X[:split], X[split:]
y_train_np, y_test_np = y[:split], y[split:]

# --- Standardize features (manual, no sklearn) ---
mean = X_train_np.mean(axis=0)
std  = X_train_np.std(axis=0) + 1e-8   # avoid division by zero
X_train_np = (X_train_np - mean) / std
X_test_np  = (X_test_np  - mean) / std

# --- Convert to PyTorch tensors ---
X_train = torch.tensor(X_train_np, dtype=torch.float32)
y_train = torch.tensor(y_train_np, dtype=torch.float32).unsqueeze(1)
X_test  = torch.tensor(X_test_np,  dtype=torch.float32)
y_test  = torch.tensor(y_test_np,  dtype=torch.float32).unsqueeze(1)

train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=32, shuffle=True)

# --- Define logistic regression model in PyTorch ---
class LogisticRegressionModel(nn.Module):
    def __init__(self, n_features):
        super().__init__()
        self.linear = nn.Linear(n_features, 1)

    def forward(self, x):
        return torch.sigmoid(self.linear(x))

model    = LogisticRegressionModel(n_features)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# --- Train ---
n_epochs = 100
for epoch in range(n_epochs):
    model.train()
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        preds = model(X_batch)
        loss  = criterion(preds, y_batch)
        loss.backward()
        optimizer.step()

# --- Evaluate ---
model.eval()
with torch.no_grad():
    y_pred_prob = model(X_test)
    y_pred = (y_pred_prob >= 0.5).float()
    accuracy = (y_pred == y_test).float().mean().item()

print(f"Accuracy: {accuracy:.4f}")
```

```python
# I used claude.ai on March 19, 2023
# Prompt: can you generate a version of this code which uses matplotlib to plot relevant data about the training and evaluation? make sure to do plt.show() at the end. [code]

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler  
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, confusion_matrix, roc_curve, auc,
    ConfusionMatrixDisplay
)
 
# --- Data Generation ---
np.random.seed(42)
n_samples = 1000
n_features = 20
X = np.random.rand(n_samples, n_features)
coefs = np.random.randn(n_features)
y = np.dot(X, coefs) > 0
y = y.astype(int)
 
# --- Preprocessing & Split ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()  
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
 
# --- Train Model ---
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
 
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
 
# ================================================================
# PLOTS
# ================================================================
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Logistic Regression — Training & Evaluation', fontsize=15, fontweight='bold')
 
# ----------------------------------------------------------------
# 1. Class Distribution (train vs test)
# ----------------------------------------------------------------
ax1 = axes[0, 0]
labels = ['Class 0', 'Class 1']
train_counts = [np.sum(y_train == 0), np.sum(y_train == 1)]
test_counts  = [np.sum(y_test  == 0), np.sum(y_test  == 1)]
x = np.arange(len(labels))
width = 0.35
ax1.bar(x - width/2, train_counts, width, label='Train', color='steelblue')
ax1.bar(x + width/2, test_counts,  width, label='Test',  color='coral')
ax1.set_title('Class Distribution')
ax1.set_xticks(x)
ax1.set_xticklabels(labels)
ax1.set_ylabel('Count')
ax1.legend()
ax1.grid(axis='y', alpha=0.4)
 
# ----------------------------------------------------------------
# 2. Confusion Matrix
# ----------------------------------------------------------------
ax2 = axes[0, 1]
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Class 0', 'Class 1'])
disp.plot(ax=ax2, colorbar=False, cmap='Blues')
ax2.set_title('Confusion Matrix')
 
# ----------------------------------------------------------------
# 3. ROC Curve
# ----------------------------------------------------------------
ax3 = axes[1, 0]
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
ax3.plot(fpr, tpr, color='steelblue', lw=2,
         label=f'ROC curve (AUC = {roc_auc:.3f})')
ax3.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=1, label='Random')
ax3.set_xlim([0, 1])
ax3.set_ylim([0, 1.02])
ax3.set_xlabel('False Positive Rate')
ax3.set_ylabel('True Positive Rate')
ax3.set_title('ROC Curve')
ax3.legend(loc='lower right')
ax3.grid(alpha=0.4)
 
# ----------------------------------------------------------------
# 4. Feature Coefficients (top 20)
# ----------------------------------------------------------------
ax4 = axes[1, 1]
coef_vals = model.coef_[0]
feature_names = [f'Feature {i+1}' for i in range(n_features)]
sorted_idx = np.argsort(np.abs(coef_vals))[::-1][:20]
sorted_coefs = coef_vals[sorted_idx]
sorted_names = [feature_names[i] for i in sorted_idx]
colors = ['steelblue' if c > 0 else 'coral' for c in sorted_coefs]
ax4.barh(range(len(sorted_coefs)), sorted_coefs[::-1], color=colors[::-1])
ax4.set_yticks(range(len(sorted_coefs)))
ax4.set_yticklabels(sorted_names[::-1], fontsize=8)
ax4.axvline(0, color='black', linewidth=0.8)
ax4.set_title('Model Coefficients')
ax4.set_xlabel('Coefficient Value')
ax4.grid(axis='x', alpha=0.4)
 
plt.tight_layout()
plt.show()
```
